GENERATORS

EX1

```python
def generate_squares(N):
    for i in range(1, N + 1):
        yield i ** 2

squares_generator = generate_squares(int(input()))

for square in squares_generator:
    print(square,end=" ")
print()
```

EX2

```python

def even(n):
    for i in range(0,n+1):
        if(i%2==0):
            yield i

n=int(input())
generator=even(n)

for i in generator:
    print(i,end="")
    if i != n and i + 2 <= n:
            print(",", end=" ")
print()
```

EX3

```python
def repeat(n):
    for i in range(0,n+1):
        if(i%3==0 and i%4==0):
            yield i

generator = repeat(int(input()))

for i in generator:
    print(i,end=" ")
print()
```

EX4

```python
def square(a,b):
    for i in range(a,b+1):
        yield i**2

squares = square(int(input()),int(input()))

for i in squares:
    print(i,end=" ")
print()
```

EX5

```python
def numbers(n):
    for i in range(0,n+1):
        yield i

generator = numbers(int(input()))

for i in generator:
    print(i,end=" ")
print()
```

```python
import math

def radian():
    degrees=int(input("Input degree: "))
    radian=(degrees*math.pi)/180
    print("Output radian:",radian)

radian()
```

EX2

```python
def  trapezoid():
    height=int(input("Height: "))
    a=int(input("First value: "))
    b=int(input("Second value: "))
    area=((a+b)*height)/2
    print("Area trapezoid:",area)

trapezoid()
```

EX3

```python
def r_polygon():
    sides=int(input("Number of sides:"))
    length_sides=int(input("Length of a side: "))
    a=length_sides/(2*math.tan(math.pi/sides))
    area=a*sides*length_sides/2
    print("Area the palygon:",area)

r_polygon()
```

```python
def parallelogram():
    length=float(input("Length: "))
    height=float(input("Height: "))
    area=length*height
    print(area)

parallelogram()
```

```python
def drop_microseconds(dt):

    dt_without_microseconds = dt.replace(microsecond=0)
    return dt_without_microseconds

if __name__ == "__main__":
    now = datetime.now()

    datetime_without_microseconds = drop_microseconds(now)

    print("Original datetime:", now)
    print("Datetime without microseconds:", datetime_without_microseconds)
```
